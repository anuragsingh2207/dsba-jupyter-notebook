```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
# Avoids multiple plt.show() calls
%matplotlib inline  
```

```python
# Univariate Visualization (Analyzing distribution of data for a single variable)
# Shows the frequency of single numerical variable 
sns.histplot(data=auto, x='price')
```

```python
#Changing the default Bin size
sns.histplot(data=auto, x='price', bins=5, color='magenta')
```

```python
#Histogram with Kernel Density Estimation
sns.histplot(data=auto, x='price', kde=True, bins=5, color='orange')
```

```python
sns.histplot(data=auto, x='curb_weight', kde=True, color='green')
```

```python
#Bivariate Analysis
sns.histplot(data=auto, x='price', hue='body_style', kde=True, color='cyan')
```

```python
g = sns.FacetGrid(auto, col="body_style")
g.map(sns.histplot, "price");
```

```python
# Shows where is the density is centered
sns.displot(auto['wheel_base'], kde = True, rug= True)
plt.show()
```

```python
# Univariate Visualization (Analyzing distribution of data for a single variable)

# displot() only supports numerical variable not categorical variable
sns.displot(auto['engine_size'])
plt.show()
```

```python
#Bivariate Visualization 
#Analyzing distribution of data for a two numerical variable
joint_plot = sns.jointplot(x=auto['engine_size'], y=auto['horsepower'])
plt.show()
```

```python


sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="hex")
plt.show()
```

```python
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="kde")
plt.show()
```

```python
# Multivariate Visualization
sns.pairplot(auto[['wheel_base','engine_size', 'horsepower']])
```

```python
# Visualization of a Categorical Variable with Numerical Variable
# Contains lots of overlapping dots without jitter not very useful
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'] , jitter=False)
```

```python
# Visualization of a Categorical Variable with Numerical Variable
# Randomly splitted dots in Jitter
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'])
```

```python
# More Cleaner Visualization compared to Stripplot
# Ordered splitted dots 
sns.swarmplot(x=auto['fuel_type'], y=auto['horsepower'] )
```

```python
sns.boxplot(x=auto['fuel_type'], y=auto['horsepower'], hue = auto['drive_wheels'] )
```

```python
# Boxplot tells the distribution of data in comparison normal distribution
# Right Skew  known as Positive Skew
# Left skew known as Negative Skew
sns.boxplot(data=auto, x ='curb_weight')
```

```python
sns.boxplot(x=auto['body_style'], y=auto['price'])
```

```python
sns.boxplot(x=auto['num-of-doors'], y=auto['horsepower'], hue = auto['fuel-type'])
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
#background top and bottom lines show confidence interval ranges
sns.lineplot(data = flights, x='month', y='passengers')
```

Confidence Interval range of time to complete the swim from A to B
At anyday 95% confidence level we can say the swimmer can same distance.

If the confidence level is increased to 99% that increases the confidence interval range.

Industry norm/Thumb Rule is 95% confidence level.

```python
#Get rid of ci
sns.lineplot(data = flights, x='month', y='passengers', ci=False)
```
