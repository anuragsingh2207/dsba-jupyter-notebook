```python
# type your code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


marks_data = { 'Name' : ['Sheldon', 'Amy', 'Priya', 'Raj'],
        'Maths' : [72, 34, 54, 76],
        'Geography': [98, 98, 78, 87]
}

marks_dataframe = pd.DataFrame(marks_data)
marks_dataframe
```

```python
# horizontal barplot   

# type your code here

marks_dataframe.plot(x='Name', y =['Maths', 'Geography'], kind='barh',rot=45)
plt.xlabel("Name")
plt.ylabel("Maths & Geography Marks")
plt.title("Student Name vs Marks")

plt.show()
```

```python
# vertical barplot   

# type your code here

marks_dataframe.plot(x='Name', y =['Maths', 'Geography'], kind='bar',rot=45)
plt.xlabel("Name")
plt.ylabel("Maths & Geography Marks")
plt.title("Student Name vs Marks")

plt.show()
```

```python
# type your code here

import plotly.graph_objs as go
import plotly.io as pio

# Sample data
x_values = [1, 22, 13, 42, 55]
y_values = [10, 12, 14, 16, 18]

# Create scatter plot
trace = go.Scatter(
    x=x_values,
    y=y_values,
    mode='markers',
    marker=dict(size=6, color='blue')
)

layout = go.Layout(title='Scatter Plot', xaxis=dict(title='X-axis'), yaxis=dict(title='Y-axis'))

fig = go.Figure(data=[trace], layout=layout)

# Save scatter plot as standalone HTML file
pio.write_html(fig, file='scatter_plot.html', auto_open=True)
```
