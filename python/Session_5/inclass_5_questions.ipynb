```python
# type your code here
#Histogram of random numbers from normal distribution

random_numbers = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(random_numbers, bins=20, density=True, alpha=0.9, color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.grid(True)
plt.show()
```

```python
#Histogram with frequency curve
random_numbers = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(random_numbers, bins=30, density=True, alpha=0.7, color='green', label='Histogram')


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
frequency = (1 / (np.sqrt(2 * np.pi) * 1)) * np.exp(-0.5 * ((x - 0) / 1) ** 2)  # PDF for normal distribution
plt.plot(x, frequency, label='Frequency Curve', color='red')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Frequency Curve for Random Numbers from Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
