Step3: Write a function determine how many times a given letter "n" occurs in a string= (" Banana tree") (5 Marks)

```python
def count_num_n(input_string):
    count = 0
    for i in input_string:
        if i == 'n':
            count +=1
    return count

print("Number of 'n's in'Banana tree' are :",count_num_n(" Banana tree"))        
```

Print that x value which satisfies this equation (5 Marks)<br>

```python
print("(c)")
for i in range(0, len(input_string)):
    count=0
    if input_string[i]=='a':    
        print("Index of 'a' is: ",i)
        count+=1
        break
if count==0:
    print("There is no 'a' in input string")
```

```python
print("(d)")
print(input_string.upper())
```

```python
user_input = input("Enter any string with atleast 6 characters: ")
count=0
print("(e)")
modified_string=""
for i in user_input:
    if i==' ':    
        modified_string+="_"
        count+=1
    else:
        modified_string+=i

if count==0:
    print("No spaces found in the entered string...")
else:        
    print("Modified String:",modified_string)
```
